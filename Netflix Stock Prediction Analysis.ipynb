### Slicing the dataset

```python
dataset=dataset.drop(['Dividends', 'Date', 'Open', 'Low', 'Volume', 'Stock Splits'],axis=1,errors='ignore')#errors='ignore' is given to ignore the key error:LOR not found in axis
```

```python
dataset
```

### Segregate and shape the variables

```python
x=dataset.iloc[:,0]
```

```python
x
```

```python
x.shape
```

```python
x=dataset.iloc[:,0].values.reshape(-1,1)#reshaping the data, so that it can be in a dataframe
```

```python
x.shape
```

```python
x
```

```python
y=dataset.iloc[:,-1].values.reshape(-1,1)
```

```python
y.shape
```

```python
y
```

### Scatter plot

```python
plt.scatter(x,y)
plt.xlabel('High')
plt.ylabel('Close')
plt.title('Daily Netflix Stock prediction')
plt.show()
```

### Divide the dataset into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Perform the linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

### Train the model-Linear Regression

```python
lm.fit(x_train,y_train)
```

### Predict the chance of admit

```python
y_pred=lm.predict(x_test)
```

```python
y_pred
```

### Checking the prediction

```python
check=pd.DataFrame(x_test,columns=['Daily High'])
```

```python
check['Actual Close']=y_test
```

```python
check['Predicted Close']=y_pred
```

```python
check
```

### Visualize the regressor line

```python
plt.scatter(x,y,color='orange')
plt.plot(x_test,y_pred,color='red')
plt.xlabel('High')
plt.ylabel('Close')
plt.title('Daily Netflix stock close prediction - Training')
```

```python
plt.scatter(x_test,y_test,color='orange')
plt.plot(x_test,y_pred,color='green')
plt.xlabel('High')
plt.ylabel('Close')
plt.title('Daily Netflix stock close prediction - Testing')
```
